```python
import numpy as np
import pandas as pd
```

```python
data=[1,2,3,4,5]
print("List:",data)
print(type(data))
```

```python
array1= np.array(data)
array2=np.array([6,7,8,9,10])
print("\nArray 1:",array1)
print("Array 2:",array2)
print(type(array1))
```

```python
mul=array1*array2
division=array2/array1
poweroff=array1**array2
print("multiplication=",mul)
print("division=",division)
print("poweroff=",poweroff)
addition=array1+array2
print("AdditionofTwo:{array1}+array2=",addition)
```

```python
import matplotlib.pyplot as plt
```
